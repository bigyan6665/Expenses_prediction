```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("insurance.csv")
df.head()
```

# 1 Data preprocessing
## 1.1 Data cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    print(df[i].unique())
```

## 1.2 Data transformation

```python
gender = pd.get_dummies(df["sex"],drop_first=True)
gender
```

```python
df["sex"].info()
```

```python
gender.info()
```

```python
regions = pd.get_dummies(df["region"])
regions
```

```python
is_smoker = pd.get_dummies(df["smoker"],drop_first=True)
is_smoker.rename(columns={"yes":"is_smoker"},inplace=True)
is_smoker
```

```python
new_df = pd.concat([df,gender,regions,is_smoker],axis=1)
new_df
```

```python
new_df.drop(["sex","smoker","region"],axis=1,inplace=True)
```

```python
new_df
```

# EDA

```python
new_df.corr()
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(data=new_df.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(data=new_df)
plt.show()
```

# 2 Modeling
## 2.1 train test split

```python
# dependent var = expenses
# independent var = all others
train_df,test_df = train_test_split(new_df,train_size=0.7,test_size=0.3,random_state=1)

indp = ["age","bmi","children","male","northeast","northwest","southeast","southwest","is_smoker"]
x_train = train_df[indp]
y_train = train_df[["expenses"]]
x_test = test_df[indp]
y_test = test_df[["expenses"]]
```

## 2.2 Model training

```python
# linear reg mmodel
lm = LinearRegression()
ln_model = lm.fit(x_train,y_train)

# polynomial reg model
p = PolynomialFeatures(degree=2)
# training
poly = LinearRegression()
poly_model = poly.fit(p.fit_transform(x_train),y_train)
```

## 2.3 Model testing

```python
# linear reg model
data = x_test.iloc[0:1]
print("Actual expense =",y_test.iloc[0:1])
print("Predicted expense =",ln_model.predict(data))
```

```python
# polynomial reg model
data = x_test.iloc[0:1]
print("Actual expense =",y_test.iloc[0:1])
print("Predicted expense =",poly_model.predict(p.fit_transform(data)))
```

# 2.4 Model evaluation

```python
plt.scatter(y_test,ln_model.predict(x_test))
plt.xlabel("y_test")
plt.ylabel("y_predicted")
plt.show()
```

```python
plt.scatter(y_test,poly_model.predict(p.fit_transform(x_test)))
plt.xlabel("y_test")
plt.ylabel("y_predicted")
plt.show()
```

```python
# training accuracy
print("linear model = ", ln_model.score(x_train,y_train)*100)
print("polynomial model = ", poly_model.score(p.fit_transform(x_train),y_train)*100)
```

```python
# testing accuracy
print("linear model = ", ln_model.score(x_test,y_test)*100)
print("polynomial model = ", poly_model.score(p.fit_transform(x_test),y_test)*100)
```

# since, polynomial model has greater accuracy while training and testing both.Thus, it is the best model

# model dumping

```python
# # #to dump model
# dumping model means saving the trained model in a file so that next time, we can directly use that model[no need to retrain]
import pickle as pk
filename = 'bestmodel.pickle'
with open(filename, 'wb') as f:
    pk.dump(poly_model,f)
```
